```python
import pandas as pd 
import numpy as np
import scipy.stats as stats 
from glob import glob
```

## 산업별 상관계수 분석과 통계적 유의미도

```python
def print_corr(file_name):
    df = pd.read_csv(file_name, index_col=0, encoding='UTF-8')
    ind_name = df['산업분류'][1]
    corr1 = stats.pearsonr(df['인구 대비 산업체 수'], df['인구 대비 확진자 수'])
    corr2 = stats.pearsonr(df['인구 대비 산업체 종사자 수'], df['인구 대비 확진자 수'])
    corr3 = stats.pearsonr(df['종사자수 대비 산업체 수'], df['인구 대비 확진자 수'])
    corr4 = stats.pearsonr(df['종사자수 대비 비정규 종사자 비'], df['인구 대비 확진자 수'])
    return [ind_name, corr1, corr2, corr3, corr4]
```

```python
ind = []
corr1 = []
corr2 = []
corr3 = []
corr4 = []

for file_name in glob('*.csv'): 
    list = print_corr(file_name)
    ind.append(list[0])
    corr1.append(list[1])
    corr2.append(list[2])
    corr3.append(list[3])
    corr4.append(list[4])
```

```python
corr_dict = {'산업명': ind, '인구 대비 산업체수': corr1, '인구 대비 산업체 종사자 수': corr2, '종사자수 대비 산업체 수': corr3,
            '종사자수 대비 비정규 종사자 비':corr4}
corr_df = pd.DataFrame(corr_dict)
corr_df
```

```python
corr_df.to_csv('상관계수 분석 결과.csv', encoding='utf-8')
```
